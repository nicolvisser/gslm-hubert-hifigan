```python
from gslm import GSLM
from torchaudio.datasets import LIBRISPEECH
from pathlib import Path
import torchaudio
import torch
```

```python
model = GSLM(n_units=50, dp_lambda=0).cuda().eval()
```

```python
librispeech_dir = Path("/home/nicolvisser/datasets/LibriSpeech")

dataset = LIBRISPEECH(
    root=librispeech_dir.parent,
    url="dev-clean",
    download=False,
    folder_in_archive=librispeech_dir.name,
)

wav, sr, *_ = dataset[0]
```

```python
with torch.inference_mode():
    units = model.encode(wav=wav.cuda(), sr=sr, dedupe=True)
    wav_, sr = model.decode(units, deduped=True)
```

```python
from IPython.display import Audio

Audio(wav_.cpu().numpy(), rate=sr)
```
